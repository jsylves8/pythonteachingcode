## Program: list-o-matic  
This program takes string input and checks if that string is in a list of strings    
- if string is in the list it removes the first instance from list  
- if string is not in the list the input gets appended to the list  
- if the string is empty then the last item is popped from the list 
- if the **list becomes empty** the program ends  
- if the user enters "quit" then the program ends  

program has 2 parts  
- **program flow** which can be modified to ask for a specific type of item.  This is the programmers choice.  Add a list of fish, trees, books, movies, songs.... your choice.  
- **list-o-matic** Function which takes arguments of a string and a list.  The function modifies the list and returns a message as seen below.

````python
# [] create list-o-matic

user_list = ["Scream", "Halloween", "The Shining", "Texas Chainsaw", "The Exorcist", "Friday The 13th"]


def list_o_matic(input_user):

    input_string = input_user

    if input_string == "":

        before_pop = user_list[-1]

        user_list.pop()

        pop_msg = (f"Here is the list of horror movies:{user_list} \nThis movie title was entered:\t{input_string} \nThe movie \"{before_pop}\" has been popped")

        return pop_msg

    else:

    
        if input_user in user_list:

            removed_word = user_list.remove(input_user)

            remove_msg = (f"Here is the list of horror movies:{user_list} \nThis movie title was entered:\t{input_string} \nThe movie \"{input_string}\" has been removed")

            return remove_msg

        else:

            append_word = user_list.append(input_user)

            append_word_str = str(append_word)

            append_msg = (f"Here is the list of horror movies:{user_list} \nThis movie title was entered:\t{input_string} \nThe movie \"{input_string}\" has been added") 

            return append_msg

           
def program_start():
    
    while True:
        
        if user_list == []:

            print(f"There is nothing in your movie list:{user_list}")
        
            break
        else:
            
            input_str = input("Please enter your favorite classic horror movie:\t")

            if input_str.lower() == "quit":
                print(f"Your current string is {user_list}")
                print("You have quit the program, restart and input words \nGoodbye!")
                break
            else:
                print(list_o_matic(input_str))

                


# [] copy and paste in edX assignment page

program_start()

````
